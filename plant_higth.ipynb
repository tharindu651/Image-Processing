```python
import cv2
import numpy as np

frame=cv2.imread(r"C:\Users\THARINDU\OneDrive\Desktop\ip\DAY 2\plants\D10.png")
hsv=cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)

lower_color=np.array((24,38,123) ) #green
upper_color=np.array((50,255,255))  #green

mask=cv2.inRange(hsv,lower_color,upper_color)

contours,hierachy=cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
#cv2.drawContours(frame,contours,-1,(0,0,0),2)


for cnt in contours:
    area=cv2.contourArea(cnt)
    #print(area)
    if (area>1500):
        cv2.drawContours(frame,[cnt],-1,(0,0,255),3)
       
        x,y,w,h = cv2.boundingRect(cnt) 
       
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0,4))
       # print(x,y,w,h)
        cv2.putText(frame,'height:'+str(h),(x+w+10,y+h),cv2.FONT_HERSHEY_SIMPLEX,0.75,(255,255,255),2)
                                             #TEXT PLACE       #FONT TYPE          
cv2.imshow('imgfa',frame)
#cv2.imshow('imgfb',hsv)
#cv2.imshow('imgfc',mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
